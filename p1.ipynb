```python
import pandas as pd

# 1. Load the raw noise
df = pd.read_excel('Dataset for Data Analytics p1.xlsx')

# 2. Phase 1: Strategic Imputation
df['CouponCode'] = df['CouponCode'].fillna('NO_COUPON')

# 3. Phase 2: Integrity Audit (Failsafe for duplicates)
df = df.drop_duplicates(subset=['OrderID'], keep='first')

# 4. Phase 3: Speak One Language
# ISO 8601 Dates
df['Date'] = pd.to_datetime(df['Date']).dt.strftime('%Y-%m-%d')

# Proper Case & Trim Whitespace
text_cols = ['Product', 'ShippingAddress', 'PaymentMethod', 'OrderStatus', 'ReferralSource']
for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.title()

# Numeric Precision (2 decimals)
df['UnitPrice'] = df['UnitPrice'].round(2)
df['TotalPrice'] = df['TotalPrice'].round(2)

# 5. Export the Production-Ready data
df.to_excel('Cleaned_Dataset_Project1.xlsx', index=False)
print("Data Cleaning Complete. File Exported.")
```
