```python
import pandas as pd

# Load the dataset using the new file name
# Note: Since this is a CSV file, we use pd.read_csv instead of pd.read_excel
df = pd.read_excel('Dataset for Data Analytics.xlsx', sheet_name='Sheet1')

# 1. Calculate basic statistics
print("--- Descriptive Statistics ---")
print(df[['Quantity', 'UnitPrice', 'ItemsInCart', 'TotalPrice']].describe().round(2))

# 2. Identify Trends: Total Sales by Product
print("\n--- Total Sales by Product ---")
sales_trend = df.groupby('Product')['TotalPrice'].sum().sort_values(ascending=False)
print(sales_trend)

# 3. Identify Outliers: Using the IQR Method on TotalPrice
Q1 = df['TotalPrice'].quantile(0.25)
Q3 = df['TotalPrice'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR

outliers = df[df['TotalPrice'] > upper_bound]
print(f"\n--- Outliers Found: {len(outliers)} ---")
print(f"Transactions exceeding the upper bound of ${upper_bound:.2f}:")
print(outliers[['OrderID', 'Product', 'Quantity', 'TotalPrice']])
```
